## 📖 Revisão rápida — O que são variáveis?

Variáveis são "caixinhas" na memória do computador que guardam informações temporárias enquanto o programa está rodando.

### 🗂️ Tipos de dados básicos no Python

| Tipo | Descrição | Exemplos |
|------|-----------|----------|
| `int` | Números inteiros | `1`, `20`, `1000`, `-5` |
| `float` | Números reais (com casas decimais) | `7.5`, `2.07`, `50.29` |
| `str` | Textos (strings) | `"Bom dia"`, `"Python"`, `"42"` |
| `bool` | Valores lógicos | `True`, `False` |

### 💡 Dicas rápidas
- Use `type()` para ver o tipo de uma variável
- `input()` **sempre** retorna `str` — converta se precisar de número!
- Para converter: `int()`, `float()`, `str()`
- Para imprimir com variáveis: `print(f"Olá, {nome}!")`

```python
# 🧪 Exemplo rápido — rode esta célula para ver como funciona!
nome = "Luke Skywalker"
ano  = 1977
nota = 9.8
jedi = True

print(f"Nome: {nome} | Tipo: {type(nome)}")
print(f"Ano:  {ano}  | Tipo: {type(ano)}")
print(f"Nota: {nota} | Tipo: {type(nota)}")
print(f"Jedi: {jedi} | Tipo: {type(jedi)}")
```
