## 📘 Tópico 2: Desvios Condicionais (if / elif / else / match)

## 📖 Revisão rápida

O **desvio condicional** permite que o programa tome decisões diferentes dependendo de uma condição.

```python
# if simples — só faz algo SE for verdadeiro
if condicao:
    print("é verdadeiro!")

# if / else — dois caminhos
if condicao:
    print("verdadeiro")
else:
    print("falso")

# if / elif / else — múltiplas condições
if nota >= 9:
    print("Ótimo!")
elif nota >= 6:
    print("Aprovado")
else:
    print("Reprovado")
```

### 🔑 Operadores de comparação
`==` igual | `!=` diferente | `>` maior | `<` menor | `>=` maior/igual | `<=` menor/igual

### 🔑 Operadores lógicos
`and` → ambos verdadeiros | `or` → pelo menos um | `not` → inverte

````python
# 🧪 Exemplo rápido — rode e observe!
pontos = int(input("Quantos pontos você tem? "))

if pontos >= 1000:
    print("🏆 Campeão! 3 GB de bônus!")
elif pontos >= 500:
    print("🥈 Muito bem! 1.5 GB de bônus!")
elif pontos >= 200:
    print("🥉 Bom! 500 MB de bônus!")
else:
    print("😕 Continue tentando, sem bônus desta vez.")
````
